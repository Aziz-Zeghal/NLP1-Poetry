## Benchmark statistics

We take weighted averages of the metrics to reflect imbalances in the dataset.

### Preparation

```python
df = benchmark.copy()
# Flatten index
df = benchmark.reset_index()
```

```python

# Melt the DataFrame to long format
df_melted = df.melt(id_vars=["Encoding", "Model"], value_vars=metrics,
                    var_name="Metric", value_name="Score")

# Optional: Create a label combining Model and Encoding if needed
df_melted["Label"] = df_melted["Encoding"] + " + " + df_melted["Model"]

def plot_metric(df, metric):
    """
    Displays a bar plot of the corresponding metric on all models.
    
    """
    # Filter for one metric, e.g., Accuracy
    df_filtered = df[df["Metric"] == metric]

    # Plot
    fig = px.bar(
        df_filtered,
        x="Model",
        y="Score",
        color="Encoding",
        barmode="group",
        title=f"{metric} by Model and Encoding",
        height=500,
        text_auto=True
    )

    fig.update_traces(textposition="inside")
    fig.update_xaxes(tickangle=45)
    fig.update_layout(
        yaxis_title=metric,
        uniformtext_minsize=8,
        uniformtext_mode="hide"
    )

    fig.show()
```

### Graphs

```python
def heat_map_plot(df: pd.DataFrame, encoding: list[str] = encodings):
    """
    Makes a plot with the associated encodings.
    """

    # Filter the DataFrame for the specified encodings
    df_filtered = df[df["Encoding"].isin(encoding)].copy()

    # Set a multi-index string label
    df_filtered["Label"] = df_filtered["Encoding"] + " + " + df_filtered["Model"]

    # Select only numeric columns
    df_metrics = df_filtered[metrics]
    df_metrics.index = df_filtered["Label"]

    plt.figure(figsize=(12, 6))
    sns.heatmap(df_metrics.drop("Avg Training time (m)", axis=1), annot=True, fmt=".2f", cmap="YlGnBu", vmin=0.69, vmax=0.99)
    plt.title("Model Performance Metrics")
    plt.ylabel("Model + Encoding")
    plt.tight_layout()
    plt.show()
```

```python
heat_map_plot(df, ["TF-IDF", "Word2Vec"])
```

```python
heat_map_plot(df, ["Word2Vec"])
```

```python
fig = px.bar(
    df_melted,
    x="Model",
    y="Score",
    color="Encoding",
    facet_col="Metric",
    barmode="group",
    title="Model Performance Metrics by Encoding",
    height=600,
    text_auto=True
)

# Rotate x-tick labels and remove repetitive titles
fig.update_xaxes(tickangle=45, title=None)

# Style improvements
fig.update_traces(textposition="inside")
fig.update_layout(
    yaxis_title="Score",
    uniformtext_minsize=8,
    uniformtext_mode="hide"
)

fig.show()
```
