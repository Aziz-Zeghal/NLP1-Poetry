# Poetry NLP
Make a model that generates poems with a title

## 🎓 Library

```python
import pandas as pd
import poetree
```

## Data retrieval

```python
corpus = poetree.Corpus('fr')

authors = corpus.get_authors()

data = [
    {
        "Titre": poem.title,
        "Texte": "".join(ligne['text'] for ligne in poem.get_body()),
        "Auteur": author.name,
        "Année de naissance": author.born if author.born else "Inconnue", 
        "Année de décès": author.died if author.died else "Inconnue"
    }
    for author in authors[:10]
    for poem in author.get_poems()
]
```
