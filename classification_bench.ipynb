Cells to run the classification benchmark.

Takes the `classification.parquet` file and makes graphs

```python
import pandas as pd
```

## Parquet file creation

```python
# Make a dataframe to contain each model's results

models = ["Logistic Regression", "Feedforward Neural Network", "Recurrent Neural Network", "Naive Bayes", "Random Forest"]
encodings = ["TF-IDF", "Word2Vec"]


# Multindex table with: Encoding then Model
index = pd.MultiIndex.from_product([encodings, models], names=["Encoding", "Model"])
benchmark = pd.DataFrame(index=index, columns=["Accuracy", "Precision", "Recall", "F1-Score", "AUC", "Avg Training time (m)"]).astype(float)
benchmark.fillna(0, inplace=True)
```

```python
benchmark
```

```python
benchmark.to_parquet("class_bench.parquet", index=True)
```
